Filter and Map

```python
# Using filter
filtered_data = df[df['How healthy do you consider yourself?'] > 3]
# using map convert numbers into percentage i.e if number 10 means 100% then calculate % of existing numbers 
# in How healthy do you consider yourself? colums and save them in new column
df['Healthy_percentages'] = df['How healthy do you consider yourself?'].map(lambda x:x*10)
```

```python
## displaying newly created column
print(df[['Healthy_percentages']].to_string(index=False))
```

```python
## Health summary based on the newly created column "Healthy_percentages"
good_health = max(df.Healthy_percentages)
bad_health  = min(df.Healthy_percentages)
average_health = np.mean(df.Healthy_percentages)
```
